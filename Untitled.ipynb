```python
from matplotlib import pyplot as plt

#matplotlib.inline
import act
import math
import xarray as xr
```

```python
def func10log10(x):
    func = lambda mydata: 10 * math.log10(mydata)
    return xr.apply_ufunc(func, x)
```
